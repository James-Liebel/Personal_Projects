```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from google.colab import files
uploaded=files.upload()

for filename in uploaded.keys():
    dataset = pd.read_csv(filename)
```

Heatmaps numerical data first

```python
numericalDataset=dataset.select_dtypes(include=['number'])

plt.figure(figsize=(8,8))
sns.heatmap(numericalDataset.corr(),
    annot=True



)
```

Finds unique values in each categorical variable.  If any = 1 then remove that column

```python
categoricalDataset=dataset.select_dtypes(include='object')
names=[]
for col in categoricalDataset:
  names.append(col)
uniqueVals=[]
for i in names:
  uniqueVals.append(dataset[i].unique().size)
plt.figure(figsize=(8,8))
sns.barplot(x=names,y=uniqueVals)
plt.show()


print(names)
```

```python
dataset.pop('OffenderStatus')
```

```python
print(dataset.columns)
```

```python
encoder=OneHotEncoder(sparse_output=False)
encoded_data=encoder.fit_transform(dataset[names])

encoded_df=pd.DataFrame(encoded_data,columns=encoder.get_feature_names_out(names))

dataset=pd.concat([dataset,encoded_df], axis=1)
dataset.drop(columns=names, inplace=True)

dataset.head()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(dataset.corr(),
    annot=True



)
```
