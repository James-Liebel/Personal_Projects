```python
import numpy as np
newData = np.array(train_dataset, dtype=np.float32)  # Ensure it's a float array
newLabels = np.array(train_labels, dtype=np.float32)  # Ensure it's a float array
model = keras.Sequential([
    keras.Input(shape=(newData.shape[1],)), # Feature layer
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(32, activation='relu'),
    keras.layers.Dense(1)  # Output layer with one neuron for regression
])

# 4. Compile the model with appropriate optimizer, loss, and metric
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])

# 5. Train the model
model.fit(newData, newLabels, epochs=5, batch_size=1)
```
