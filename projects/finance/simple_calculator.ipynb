```python
import sympy
from sympy import symbols
from sympy.solvers import solve
import math
from fractions import Fraction
import matplotlib.pyplot as plt
import numpy as np

# Write your code here



def prop():
  n1=int(input('n1='))
  d1=int(input('d1='))
  n2=int(input('n2='))
  d2=int(input('d2='))

  if n1==0:
    return n2*d1/d2
  elif n2==0:
    return n1*d2/d1
  else:
    return 'make numerator a variable'

def solve_x():
  x=symbols('x')
  eq=input('equation:')
  return solve(eq,x)


def fac_roots():
  num=int(input('number to root:'))

  max=math.floor(math.sqrt(num)+1)
  big_fac=1

  for nums in range(1,max):
    if num%(nums**2)==0:
      big_fac=nums

  if big_fac==1:
    return sympy.sqrt(num)
  else:
    o_num=num//(big_fac**2)
    return big_fac*sympy.sqrt(o_num)

def dec_to_frac():
  dec=float(input('enter decimal:'))
  decision=input('do you want to make it a fraction(f) or percent(p)?:')

  if decision=='f':

    return str(Fraction(dec).limit_denominator())
  elif decision=='p':
    return str(round(dec*100,2))+' percent'
  else:
    return "you typed something wrong"



def frac_to_dec():
  n=int(input('numerator:'))
  d=int(input('denominator:'))
  decision=input('do you want to make it a decimal(d) or percent(p)?:')

  num=float(n/d)
  if decision == 'd':
    return num
  elif decision == 'p':
    return round(num*100,2)
  else:
    return "you typed something wrong"

def per_to_frac():
  num=float(input('give input as jsut a decimal number:'))*100

  decision =input('do you want to make it a fraction(f) or a decimal(d)?:')

  if decision== 'f':
    return str(Fraction(num).limit_denominator())
  elif decision=='d':
    return num
  else:
    return "you typed something wrong"

def other():
  func=input('function you want to graph:')


  xmin = -10
  xmax = 10
  ymin = -10
  ymax = 10
  points = 2*(xmax-xmin)
  x = np.linspace(xmin, xmax, points)

  fig, ax = plt.subplots()
  plt.axis([xmin,xmax,ymin,ymax]) # window size
  plt.plot([xmin,xmax],[0,0],'b') # blue x axis
  plt.plot([0,0],[ymin,ymax], 'b') # blue y axis

  ax.set_xlabel("x values")
  ax.set_ylabel("y values")
  ax.set_title("Some Graph")
  ax.grid(True)

  ax.set_xticks(np.arange(xmin, xmax, 1))
  ax.set_yticks(np.arange(ymin, ymax, 1))

  y=eval(f'lambda x: {func}')
  y_vals=y(x)
  plt.plot(x,y_vals,label=str(func))

  ax.legend()
  plt.show()


# This step does not have test
```
