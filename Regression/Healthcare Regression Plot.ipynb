```python
# Import data
!wget https://cdn.freecodecamp.org/project-data/health-costs/insurance.csv
dataset = pd.read_csv('insurance.csv')
dataset.tail()
```

```python
import tensorflow as tf

# Assuming dataset is already defined and is a Pandas DataFrame
feature_columns = [
    tf.feature_column.categorical_column_with_vocabulary_list("sex", ["male", "female"]),
    tf.feature_column.categorical_column_with_vocabulary_list("smoker", ["yes", "no"]),
    tf.feature_column.categorical_column_with_vocabulary_list("region", ["southeast", "southwest", "northeast", "northwest"]),
    tf.feature_column.numeric_column("age"),
    tf.feature_column.numeric_column("bmi"),
    tf.feature_column.numeric_column("children"),
    tf.feature_column.numeric_column("expenses")
]

# Split data into two lists
split_index = int(0.8 * len(dataset))
train_dataset = dataset[:split_index]
test_dataset = dataset[split_index:]

train_labels = train_dataset.pop('expenses')
test_labels = test_dataset.pop('expenses')

model = keras.Sequential([
    tf.keras.layers.Dense(feature_columns),  # Feature layer
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(32, activation='relu'),
    keras.layers.Dense(1)  # Output layer with one neuron for regression
])

# 4. Compile the model with appropriate optimizer, loss, and metric
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])

# 5. Train the model
model.fit(train_dataset, train_labels, epochs=10, batch_size=32)




#def make_input_fn(features, labels, num_epochs=10, shuffle=True, batch_size=32):
    #def input_function():
        #ds = tf.data.Dataset.from_tensor_slices((dict(features), labels))
        #if shuffle:
            #ds = ds.shuffle(100)
        #ds = ds.batch(batch_size).repeat(num_epochs)
        #return ds
    #return input_function

# Create input functions
#train_input_fn = make_input_fn(train_dataset, train_labels)
#eval_input_fn = make_input_fn(test_dataset, test_labels, num_epochs=1, shuffle=False)


# Create the estimator

```
