```python
# Import data
!wget https://cdn.freecodecamp.org/project-data/health-costs/insurance.csv
dataset = pd.read_csv('insurance.csv')
dataset.tail()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.ensemble import RandomForestRegressor

categorical_cols=['sex','smoker','region']

encoder=OneHotEncoder(sparse_output=False)
encoded_data=encoder.fit_transform(dataset[categorical_cols])

encoded_df=pd.DataFrame(encoded_data,columns=encoder.get_feature_names_out(categorical_cols))

dataset=pd.concat([dataset,encoded_df], axis=1)
dataset.drop(columns=categorical_cols, inplace=True)

dataset.head()

labels=dataset.pop('expenses')

train_dataset,test_dataset,train_labels,test_labels=train_test_split(dataset,labels,test_size=0.2,random_state=42)

rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(train_dataset, train_labels)
#model=LinearRegression()
#model.fit(train_dataset,train_labels)

test_pred=rf_model.predict(test_dataset)


```

```python
plt.scatter(test_labels,test_pred)
plt.xlabel('actual numbers')
plt.ylabel('predicted numbers')
plt.title('predicted vs actual nums of expenses')
plt.show()
```

```python
# Calculate predictions
test_predictions = rf_model.predict(test_dataset)

# Compute evaluation metrics
mae = mean_absolute_error(test_labels, test_predictions)
mse = mean_squared_error(test_labels, test_predictions)

# Print evaluation results
print("Testing set Mean Abs Error: {:5.2f} expenses".format(mae))

if mae < 3500:
    print("You passed the challenge. Great job!")
else:
    print("The Mean Abs Error must be less than 3500. Keep trying.")

# Plot predictions
plt.figure(figsize=(8, 6))
a = plt.axes(aspect='equal')
plt.scatter(test_labels, test_predictions, alpha=0.5)
plt.xlabel('True values (expenses)')
plt.ylabel('Predictions (expenses)')
lims = [0, 50000]
plt.xlim(lims)
plt.ylim(lims)
_ = plt.plot(lims, lims, 'k--', lw=2)
plt.title('True vs Predicted Expenses')
plt.show()
```
