```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
uploaded=files.upload()
#euro-daily-hist_1999_2022.csv
```

```python
exchange_rates=pd.read_csv(list(uploaded.keys())[0])
exchange_rates.tail()
```

```python

def countries(countries):
  index_values=[]
  for i in exchange_rates['Period\\Unit:']:
    index_values.append(pd.to_datetime(i))

  fig, ax=plt.subplots(figsize=(20,10))
  for i in countries:
    money=pd.to_numeric(exchange_rates[i],errors='coerce')
    ax.plot(index_values,money,label=f'{i}')
  ax.set(xlabel='index', ylabel='us dollar values', title='exchange rates over time')
  ax.legend()
  plt.show()
```

```python
countryList=['[Canadian dollar ]','[Australian dollar ]','[Singapore dollar ]']
countries(countryList)
```
