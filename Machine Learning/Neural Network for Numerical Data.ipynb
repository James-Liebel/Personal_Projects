```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from google.colab import files
uploaded=files.upload()

for filename in uploaded.keys():
    dataset = pd.read_csv(filename)
```

```python
numericalDataset=dataset.select_dtypes(include=['number'])

plt.figure(figsize=(8,8))
sns.heatmap(numericalDataset.corr(),
    annot=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from tensorflow import keras
from tensorflow.keras import layers
import tensorflow as tf
```

```python
labels=dataset.pop('Outcome')
```

```python
train_data,test_data,train_labels,test_labels=train_test_split(dataset,labels,test_size=0.2,random_state=34)


newData = np.array(train_data, dtype=np.float32)  # Ensure it's a float array
newLabels = np.array(train_labels, dtype=np.float32)  # Ensure it's a float array
model = keras.Sequential([
    keras.Input(shape=(newData.shape[1],)), # Feature layer
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(32, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')  # Output layer with one neuron for regression
])

# 4. Compile the model with appropriate optimizer, loss, and metric
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# 5. Train the model
model.fit(newData, newLabels, epochs=50, batch_size=32)
```

Scatter plot models the predictions use the next box for a percentage if this doesn't apply well

```python
#testDataTwo = np.array(test_data, dtype=np.float32)  # Replace with your actual test dataset
prediction = model.predict(test_data)

plt.scatter(test_labels,prediction)
plt.xlabel('actual numbers')
plt.ylabel('predicted numbers')
plt.title('predicted vs actual nums of expenses')
b, a = np.polyfit(test_labels, prediction, deg=1)
xseq = np.linspace(min(test_labels), max(test_labels), num=100)
plt.plot(xseq, a + b * xseq, color="k", lw=2)
plt.show()
```
