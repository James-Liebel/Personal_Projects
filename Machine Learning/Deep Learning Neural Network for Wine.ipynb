```python
# Import Required Libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Read in white wine data
white = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv", sep =';')

# Read in red wine data
red = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv", sep =';')
```

```python
import tensorflow as tf
from sklearn.model_selection import train_test_split
import seaborn as sns
```

```python
def heatmap(dataset):
 plt.figure(figsize=(10,10))
 sns.heatmap(dataset.corr(),annot=True)
heatmap(white)
heatmap(red)
```

```python
red_labels=red.pop('quality')
white_labels=white.pop('quality')
```

```python
red_train_data,red_test_data,red_train_labels,red_test_labels=train_test_split(red, red_labels, test_size=.2, random_state=43)
white_train_data,white_test_data,white_train_labels,white_test_labels=train_test_split(white, white_labels, test_size=.2, random_state=43)
```

```python
from tensorflow import keras
from tensorflow.keras import layers

model = keras.Sequential([
    keras.Input(shape=(11,)), # Feature layer
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(32, activation='relu'),
    keras.layers.Dense(16,activation='relu'),
    keras.layers.Dense(1)  # Output layer with one neuron for regression
])

model.compile(optimizer='adam',loss='mean_squared_error')
model.fit(red_train_data,red_train_labels,batch_size=3,epochs=8)
```

```python
prediction=model.predict(red_test_data)
prediction = prediction.reshape(-1, 1)  #
prediction_df=pd.DataFrame(prediction,columns=['predicted quality'])
rounded_data=np.round(prediction)
final_data=pd.DataFrame(rounded_data,columns=['predicted quality'])
```
