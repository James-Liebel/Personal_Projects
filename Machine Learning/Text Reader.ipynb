```python
# get data files
!wget https://cdn.freecodecamp.org/project-data/sms/train-data.tsv
!wget https://cdn.freecodecamp.org/project-data/sms/valid-data.tsv

train_file_path = "train-data.tsv"
test_file_path = "valid-data.tsv"
```

```python
train_data=pd.read_csv(train_file_path,sep='\t',names=['label','message'])
train_data['label']=train_data['label'].map({'ham':0,'spam':1})

test_data=pd.read_csv(test_file_path,sep='\t',names=['label','message'])
test_data['label']=train_data['label'].map({'ham':0,'spam':1})

```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline
```

```python
x_train=train_data['message']
y_train=train_data['label']

model=make_pipeline(CountVectorizer(),MultinomialNB())
model.fit(x_train,y_train)
```

```python
# function to predict messages based on model
# (should return list containing prediction and label, ex. [0.008318834938108921, 'ham'])
def predict_message(pred_text):

  prediction_prob=model.predict_proba([pred_text])[0]
  likelihood=prediction_prob[1]
  label='spam' if likelihood >0.5 else 'ham'

  return [likelihood,label]

pred_text = "how are you doing today?"

prediction = predict_message(pred_text)
print(prediction)
```
