```python
import pandas as pd
import numpy as np
from google.colab import files
import io
import zipfile
import tensorflow as tf
import matplotlib.pyplot as plt

```

```python
def data():
  decider=input('do you want to type url (u) or upload a file (f):')
  if decider=='u':
    url = input("type url here:")
    df=pd.read_csv(url)
  elif decider =='f':
        try:
            from google.colab import files
            uploaded = files.upload()

            # Get the uploaded file name (it's a ZIP file)
            filename = list(uploaded.keys())[0]

            # Read the ZIP file
            with zipfile.ZipFile(io.BytesIO(uploaded[filename]), 'r') as zip_ref:
                # Extract the CSV file from the ZIP (assuming it has only one file)
                csv_filename = zip_ref.namelist()[0]  # Get the first file in the ZIP
                with zip_ref.open(csv_filename) as csv_file:
                    df = pd.read_csv(csv_file)
        except Exception as e:
            print(f"Error reading the file: {e}")
  else:
    #can type it in the code here
    url =''
    df=pd.read_csv(url)
  what_next=input('want to make a scatter plot(y)?:')
  if what_next=='y':
    print(df.columns)
    what_column=input('first column:')
    what_column_two=input('second column:')
    manage_data(df,what_column,what_column_two)


def manage_data(dataframe,column_one,column_two):

  plt.scatter(pd.to_numeric(dataframe[column_one],errors='coerce'),dataframe[column_two].str.replace(',', '').astype(float))
  plt.show()

```
